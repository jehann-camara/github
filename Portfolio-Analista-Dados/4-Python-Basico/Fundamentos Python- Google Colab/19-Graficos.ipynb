```python
# Importar bibliotecas para criação de gráficos
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,2,100)  # Esta função retorna 100 valores no intervalo entre 0 e 2
```

```python
x
```

```python
y = 2*x  # definição das coordenadas de y do plano cartesiano, para assim ter o conjunto de pares ordenados, para ser feita a "plotagem"
```

```python
plt.plot(x,y,"*") # Esta Função faz a plotagem dos pares ordenados no plano cartesiano
```

```python
plt.plot(x,y,"*") # Esta Função faz a plotagem dos pares ordenados no plano cartesiano
plt.show()
```

```python
plt.plot(x,y,"*") # Esta Função faz a plotagem dos pares ordenados no plano cartesiano
plt.xlabel("Eixo x - Velocidade")
plt.ylabel("Eixo y - Aceleração")
plt.title("Título- Velocidade x Aceleração")
plt.show()
```

```python
y2 = x**2 # Parábola
plt.plot(x,y2,"red") # Esta Função faz a plotagem dos pares ordenados no plano cartesiano / terceiro argumento "Y" = yellow, r = red etc
plt.xlabel("Eixo x - Velocidade")
plt.ylabel("Eixo y - Aceleração")
plt.title("Título- Velocidade x Aceleração")
plt.show()
```

```python
# executar sub-plots

#Funções
x = np.linspace(-1,1,100)
```

```python
y1 = x*2
y2 = x**2
y3 = x**3
y4 = np.sin(x)
```

```python
# subplots
fig , janela = plt.subplots(2,2)
```

```python
fig , janela = plt.subplots(3,2)
```

```python
fig , janela = plt.subplots(2,2)
janela[0,0].plot(x,y1,'blue')
janela[0,1].plot(x,y2,'red')
janela[1,0].plot(x,y3,'green')
janela[1,1].plot(x,y4,'yellow')
fig.legend(['Reta','Parábola','Cúbica', "Sin" ])
plt.show()
```
