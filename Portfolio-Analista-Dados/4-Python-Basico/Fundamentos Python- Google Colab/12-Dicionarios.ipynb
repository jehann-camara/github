```python
familiaApto101 = {"Jehann":35,"Suérllen":37,"André":True,"Socorro":61,"Kora":{"Idade":2, "Cor":"Caramelo"}}
```

```python
familiaApto101["Kora"]
```

```python
familiaApto101["Kora"]["Cor"] # aninhamento usando a chave de um dicionario dentro de outro e sua respectuva chave externa.
```

```python
# Casamento =
tamanhofamiliaAntes = len(familiaApto101) 
familiaApto101.pop("Suérllen") # Método POP exclui elementos do dicionário
familiaApto101.pop("André")
tamanhofamiliaDepois = len(familiaApto101)
```

```python
TodosPetistas = tamanhofamiliaDepois < tamanhofamiliaAntes
```
