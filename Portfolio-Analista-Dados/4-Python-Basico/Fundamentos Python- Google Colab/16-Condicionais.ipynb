```python
# 
A = int(input("Digite o valor de A: ")) # Solicita um valor para Input um int para a variável A
B = int(input("Digite o valor de B: "))
```

```python
if (A > B): # "if" equivale a condicional lógica "Se", ":" equivale a "então"
  print("A é maior do que B")

if(A <= B):
  print("A é menor ou igual que B")
```

```python
# "Else" equivale a condicional lógica "Se não"
if (A > B):
  print("A é maior do que B")
else:
  print("A é menor ou igual que B")
```

```python
#  Condicional "elif" equivalente a "se não, se "

nota = 9
```

```python
if nota <= 1 and nota > 0:
  print("Nota muito baixa")
elif nota > 1 and nota <= 4:
    print("Nota baixa.")
elif nota > 4 and nota <= 7:
    print("Nota na média.")
elif nota > 7 and nota <= 8:
    print("Nota alta!")
else:
    print("Nota inválida!")
```
