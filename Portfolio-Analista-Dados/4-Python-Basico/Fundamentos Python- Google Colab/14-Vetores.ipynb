```python
# Vetores ou Arrays
# Diferente de lista, dicionario e tuplas, os arryas tem que possuir
# os elementos todos do mesmo tipo

import numpy as np
```

```python
vetor = np.array([1,2,3,4,5,6]) # atenção usa parenteses e colchetes juntos
```

```python
type(vetor)
```

```python
vetor[4]
```

```python
lista = [1,2,3,4,5]
```

```python
vetor = np.array(lista) # transforma lista em vetor
```

```python
vetor
```

```python
type(vetor[2])
```

```python
lista2 = [1,2,3,4,5, "6"] # tipos de elementos diferentes
```

```python
vetor2 = np.array(lista2) 
# transforma lista em vetor,
# com todos elementos de um único tipo, no caso em strings
vetor2
```

```python
type(vetor2[5]) # retorna o tipo string
```

```python
vetor3 = np.array([1,2,3,4,5,"a"]) 
# Neste caso fica claro que 
# é mais complicado transformar um tipo string em inteiro do que o inverso,
# oq explica o python transformar todos em strings.
```
