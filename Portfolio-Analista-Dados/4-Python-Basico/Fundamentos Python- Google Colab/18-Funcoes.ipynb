```python
# Funcoes que retornam algum dado, variavel, retorno etc

def potencia(x):
  p = x**2   #Variavel local ( dentro da função)
  return p
```

```python
potencia(2)
```

```python
potencia(5)
```

```python
potencia(0)
```

```python
def potencia_n2(x,n=2):
   p = x**n   #Variaveis locais ( dentro da função)
   return p
```

```python
potencia_n2(6)
```

```python
potencia_n2(8)
```

```python
potencia_n2(2,6)
```

```python
# funções e seus retornos

def soma(a,b):
  s = a + b
  return s
```

```python
soma(7,5)
```

```python
def multiplicação(a,b):
  m = a * b
  return m
```

```python
multiplicação(0,14)
```

```python
def soma_multiplica(a,b):
  s = soma(a,b)
  m = multiplicação(a,b)
  return print("Soma = " + str(s) + ", Multiplicação = " + str(m))
```

```python
soma_multiplica(4,5)
```

```python
x = soma_multiplica(7,12)
```

```python
type(x)
```

```python
def soma_multiplica_str(a,b):
  s = soma(a,b)
  m = multiplicação(a,b)
  st = "Soma = " + str(s) + ", Multiplicação = " + str(m)
  return st
```

```python
soma_multiplica_str(13,5)
```

```python
y = soma_multiplica_str(5,13)
```

```python
y
```

```python
type(y)
```

```python


# Funções com argumentos indeterminados

def soma_valores(*valores):
  soma = 0
  for i in valores:
    soma = soma + i
  return soma  # atenção a identação de return para não estar dentro do for

  
```

```python
soma_valores(5,6,7,8,9)
```
