```python
# Multiindex
import pandas as pd
arrays = [[1,1,2,2],["red", "Blue","red", "Blue"]]

v = pd.MultiIndex.from_arrays(arrays, names = ("Números", "Cores"))
v
```

```python
v.codes
```

```python
v.levels
```

```python
# Multiindex - Produto Cartesiano
numeros = [0,1,2,3]
cores = ["azul", "verde", "vermelha", "amarela", "branca"]
v2 = pd.MultiIndex.from_product([numeros,cores], names= ["number","Color"])
v2
```

```python
v2.codes
```

```python
v2.levels
```
