```python
# Função Pivot Table // By Jehann Câmara
import pandas as pd
```

```python
carros = [7, 4, 3, 2, 8]
dias = pd.date_range('20250101', "20250101", periods=5)
vendedores = ['George', 'Vagner', 'Pedro', 'Vagner', 'George']

df = pd.DataFrame({'Carros Vendidos': carros, 'Vendedor': vendedores, 'Data': dias})
df
```

```python
pd.pivot_table(df, index=['Data'], columns='Vendedor', values='Carros Vendidos',)
```

```python
# Pivot Table - Explicação Didática
'''
A função pivot_table é uma ferramenta poderosa para reorganizar e resumir dados:

1. Estrutura básica:
pd.pivot_table(data, values, index, columns, aggfunc)

2. Parâmetros principais:
- data: DataFrame fonte
- values: Coluna(s) que queremos analisar
- index: Define as linhas da tabela
- columns: Define as colunas da tabela
- aggfunc: Função de agregação (sum, mean, count, etc.)

Exemplos práticos com nosso DataFrame:
'''

# Exemplo 1: Total de carros por vendedor
print("Exemplo 1: Total de carros por vendedor")
print(pd.pivot_table(df, 
                    values='Carros Vendidos',
                    index='Vendedor',
                    aggfunc='sum'))

print("\nExemplo 2: Média de carros por vendedor")
print(pd.pivot_table(df, 
                    values='Carros Vendidos',
                    index='Vendedor',
                    aggfunc='mean'))

print("\nExemplo 3: Múltiplas métricas por vendedor")
print(pd.pivot_table(df, 
                    values='Carros Vendidos',
                    index='Vendedor',
                    aggfunc=['sum', 'mean', 'count']))
```
