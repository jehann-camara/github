```python
import pandas as pd
import numpy as np
```

```python
# Exemplo de DataFrame
datas = pd.date_range("20250101", periods=6)
df = pd.DataFrame(np.random.randn(6,4), index=datas, columns=["Var_A","Var_B","Var_C","Var_D"])
df
```

```python
# Transpor Linhas em Colunas e vice-versa
df_transposto = df.T
df_transposto
```

```python
# Priimeiro linhas = 4 , depois colunas = 6
df_transposto.shape
```

```python
# Extrair apenas os valores das linhas e colunas
valores = df_transposto.values
valores
```

```python
# Size de Valores
np.size(valores)
```

```python
# A multiplicação entre os parametros de reshape tem que ser igual ao total do size de valores
valores.reshape(2,12)
```
