```python
import pandas as pd

# Stack e Unstack // By Jehann Câmara
# Importando para um dataframe dados de um arquivo CSV via Link.
df = pd.read_csv("https://cdncontribute.geeksforgeeks.org/wp-content/uploads/nba.csv")
df.head(10)
```

```python
# Verificando o tamanho/shape do DataFrame / 458 linhas(Jogadores) e 9 colunas
df.shape
```

```python
# Função Stack
stacked_df = df.stack()
stacked_df.head(10)
```

```python
# Função Unstack
unstacked = stacked_df.unstack()
unstacked.head(10)
```

```python
# Resumo didático das funções read_csv, stack e unstack com exemplos de codigo e comentarios

# 1. read_csv()
print("1. Exemplo de read_csv:")
# df já foi criado anteriormente com:
df = pd.read_csv("https://cdncontribute.geeksforgeeks.org/wp-content/uploads/nba.csv")
print(df.head(2))  # Mostrando as 2 primeiras linhas

# 2. stack()
print("\n2. Exemplo de stack:")
# Demonstração com uma amostra pequena
amostra = df.head(2)
stacked_amostra = amostra.stack()
stacked_amostra
print(stacked_amostra)

# 3. unstack()
print("\n3. Exemplo de unstack:")
# Usando o stacked_amostra anterior
unstacked_amostra = stacked_amostra.unstack()
print(unstacked_amostra)


# Resumindo:
# - read_csv: lê dados de CSV para DataFrame.
# - stack: transforma colunas em linhas (formato longo).
# - unstack: transforma linhas (índices internos) em colunas (formato largo).
```
