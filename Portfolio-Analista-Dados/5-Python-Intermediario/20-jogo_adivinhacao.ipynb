```python

# Jogo adivinha√ß√£o // By Jehann C√¢mara
# Regras:
'''As regras s√£o bastante simples. Vamos a elas:
- O jogador dever√° escolher um n√∫mero inteiro qualquer entre 0 e 50;
- O jogador receber√° dicas se seu chute est√° acima ou abaixo do valor oculto
gerado pelo computador de maneira aleat√≥ria, caso n√£o acerte de primeira o
n√∫mero;
- O jogador ter√° 10 tentativas para adivinhar qual foi o n√∫mero oculto gerado
pelo computador entre o intervalo (0 a 50).
- Informe ao jogador o n√∫mero de tentativas feitas e quantas tentativas ainda restam
para ele.'''
```

```python


# Inicializa√ß√£o
# O padr√£o do random.randint(a, b) √© INCLUIR ambos os extremos.
# Para n√∫mero entre 0 e 50, usamos (0, 50).
#numero_oculto = random.randint(0, 50) 

import numpy as np
numero_randomico = np.random.default_rng()
numero_oculto = numero_randomico.integers(50) # N√∫mero randomico entre 0 e 50

print("Escolha um n√∫mero entre 0 e 50. Voc√™ tem 10 tentativas para acertar o n√∫mero oculto.")

tentativas_restantes = 10
palpite_correto = False

# Usamos um loop WHILE para controlar explicitamente as tentativas restantes
while tentativas_restantes > 0 and not palpite_correto:
    
    try:
        # A fun√ß√£o input() aqui se refere √† fun√ß√£o do IPython/Jupyter,
        # que deve renderizar um campo de entrada na c√©lula.
        x_str = input(f"Tentativa {11 - tentativas_restantes} de 10. Digite seu palpite: ")
        x = int(x_str)
        
        if x == numero_oculto:
            print(f"\nüéâ Parab√©ns, voc√™ acertou o n√∫mero oculto que √©: {numero_oculto}. Fim de Jogo!")
            palpite_correto = True
            
        else:
            tentativas_restantes -= 1 # Decrementa apenas se errou
            
            if tentativas_restantes == 0:
                print("\nüòî Tentativas esgotadas, fim de Jogo !!")
            elif x < numero_oculto:
                print(f"O n√∫mero oculto √© MAIOR que {x}. Tente novamente. Restam: {tentativas_restantes} tentativas.")
            else: # x > numero_oculto
                print(f"O n√∫mero oculto √© MENOR que {x}. Tente novamente. Restam: {tentativas_restantes} tentativas.")
                
    except ValueError:
        # Se o usu√°rio digitar algo que n√£o √© um n√∫mero, informamos e o loop continua, gastando uma tentativa.
        print("‚ö†Ô∏è Entrada inv√°lida! Por favor, digite apenas n√∫meros inteiros. Esta tentativa foi gasta.")
        tentativas_restantes -= 1
        
# Bloco final se o loop terminou porque tentativas_restantes chegou a zero
if not palpite_correto and tentativas_restantes == 0:
    print(f"O n√∫mero secreto era: {numero_oculto}.")
```
