```python
# Função Pivot // By Jehann Câmara
import pandas as pd
import numpy as np
```

```python
dias = pd.date_range("20250101", periods=12)
dias
```

```python
pessoas = ["Jehann", "Jeann", "Jacks filho", "Suérllen", "Karyne"] 
np.random.choice(pessoas)
```

```python
nome = []
gasto = []
for i in range(12):
    nome.append(np.random.choice(pessoas))
    gasto.append(np.round(np.random.rand() * 100,2))

nome
```

```python
gasto
```

```python
# Criando DataFrame
df = pd.DataFrame({"Dia": dias, "Nome": nome, "Gasto": gasto})
df
```

```python
# Função Pivot
df_pivot = df.reset_index().pivot(index='Dia', columns='Nome', values='Gasto')  
df_pivot
```

```python
# Resumo da Operação Pivot:
print("Resumo da Operação Pivot:\n")

print("1. DataFrame Original (df):")
print("- Mostra transações individuais por data, pessoa e valor")
print("- Cada linha representa uma transação única\n")

print("2. DataFrame Pivotado (df_pivot):")
print("- Reorganiza os dados para mostrar gastos por pessoa ao longo das datas")
print("- Cada coluna representa uma pessoa")
print("- Cada linha representa uma data")
print("- Valores NaN indicam que não houve gasto naquela data\n")

# Cálculo de métricas úteis
total_por_pessoa = df_pivot.sum().round(2)
print("Total de gastos por pessoa:")
print(total_por_pessoa)
print("\nNúmero de transações por pessoa:")
print(df_pivot.count())
print("\nMédia de gastos por transação:")
print((total_por_pessoa / df_pivot.count()).round(2))

# Exemplo de como usar pivot:
print("\nDica: A função pivot é útil quando você precisa:")
print("1. Transformar valores únicos em colunas")
print("2. Visualizar dados em formato de matriz")
print("3. Analisar tendências ao longo do tempo por categoria")
```
