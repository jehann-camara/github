```python
# Merge e GroupBy // By Jehann Câmara
import pandas as pd
```

```python
# Cadastro da loja A
cadastro_a = {  
                "Id":    ["AA2648","BB2233", "CC1675", "DD2497","EE2198"],
                "Nome":  ["Aya", "Cesar", "Bruna", "Maria", "Cristina"],
                "Idade": [1,28,27,35,45],
                "CEP":   ["60125-050", "60125-040","60125-030", "60125-020", "60125-010"]
             }

cadastro_a = pd.DataFrame(cadastro_a, columns = ["Id","Nome","Idade","CEP"]) 
cadastro_a 
```

```python
# Cadastro da loja B
cadastro_b = {  
                "Id":    ["AA3448","BB2233", "CC1675", "DD2347","EE2578"],
                "Nome":  ["Bruno", "Cesar", "Cristiano", "Ronaldo", "Julio"],
                "Idade": [19,28,37,45,60],
                "CEP":   ["60135-050", "60135-040","60135-030", "6015-020", "60125-010"]
             }

cadastro_b = pd.DataFrame(cadastro_b, columns = ["Id","Nome","Idade","CEP"]) 
cadastro_b 
```

```python
# Registro de compras em todas as Lojas
compras = {  
                "Id":    ["CC1675","BB2233", "BB2233", "DD2347","AA2648", "BB2353", "CC3875", "EE2198"],
                "Data":  ["2025-01-01", "2024-01-01", "2023-01-01", "2022-01-01", "2021-01-01", "2026-01-01", "2019-01-01", "2018-01-01"],
                "Valor": [200,247,50,30,150,100, 45,60]
             }

compras = pd.DataFrame(compras, columns = ["Id","Data","Valor"]) 
compras
```

```python
# Merge
#pd.merge(tabela_esquerda, tabela_direita, on="coluna coincidente", how"left|rigth|inner|outer)

pd.merge(cadastro_a, cadastro_b, on="Id", how="inner")
```

```python
# Filto por tabela B/y - ["Id", "Idade", "CEP"]
pd.merge(cadastro_a, cadastro_b[["Id", "Idade", "CEP"]], on="Id", how="left")
```

```python
# Filto por tabela y - ["Id", "Idade", "CEP"]
pd.merge(cadastro_a, cadastro_b[["Id", "Idade", "CEP"]], on="Id", how="left" , suffixes=("_A","_B"))
```

```python
#Full Join ( contém duplicados)
lojas = pd.concat([cadastro_a,cadastro_b], ignore_index=True)
lojas
```

```python
# Remover duplicados/ clientes únicos
lojas.drop_duplicates(subset="Id")
```

```python
#Left Join
pd.merge(cadastro_a, compras, on="Id", how="left")
```

```python
# GroupBY
import numpy as np
import pandas as pd

df4 = pd.DataFrame({  
                "A": ["verdadeiro","falso","verdadeiro","falso","verdadeiro","falso","verdadeiro","falso"],
                "B": ["um","dois","um","tres","tres","um","tres","dois"],
                "C": np.random.randn(8),
                "D": np.random.randn(8)
             })

df4
```

```python
df4.groupby(["A"]).sum() # Agrupamento com soma
```

```python
df4.groupby(["B"]).sum() # Agrupamento com soma
```

```python
df4.groupby(["A", "B"]).sum() # Agrupamento com soma agrupando por A e B
```
