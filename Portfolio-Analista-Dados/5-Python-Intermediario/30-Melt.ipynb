```python
import pandas as pd

# Criando um DataFrame de exemplo para demonstrar o melt
df = pd.DataFrame({
    'Nome': ['João', 'Maria', 'Pedro'],
    'Nota_Mat': [8.5, 7.0, 9.0],
    'Nota_Port': [7.5, 8.0, 6.5],
    'Nota_Hist': [9.0, 8.5, 7.0]
})

# Mostrando o DataFrame original
print("DataFrame Original:")
print(df)
print("\n")

# Exemplo 1: Transformando todas as colunas de notas em linhas
# O melt transforma colunas em linhas, útil para análise de dados
df_melted = pd.melt(df, 
                    id_vars=['Nome'],  # Coluna que manteremos como identificador
                    value_vars=['Nota_Mat', 'Nota_Port', 'Nota_Hist'],  # Colunas que serão transformadas
                    var_name='Disciplina',  # Nome da nova coluna com os nomes das antigas colunas
                    value_name='Nota')  # Nome da nova coluna com os valores

print("DataFrame após melt (todas as notas):")
print(df_melted)
print("\n")

# Exemplo 2: Usando melt apenas com algumas disciplinas
df_melted_parcial = pd.melt(df,
                           id_vars=['Nome'],
                           value_vars=['Nota_Mat', 'Nota_Port'],  # Apenas Matemática e Português
                           var_name='Disciplina',
                           value_name='Nota')

print("DataFrame após melt (apenas Mat e Port):")
print(df_melted_parcial)
```
